Penyelesaian Permasalahan Integer Knapsack menggunakan Dynamic Proggramming

```python
import pandas as pd
import time

# Fungsi untuk menyelesaikan masalah Integer Knapsack menggunakan Dynamic Programming
def knapsackDP(weights, profits, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Mengisi tabel DP
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if int(weights[i - 1]) <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][int(w - weights[i - 1])] + profits[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    # Menentukan item yang dipilih
    total_profit = dp[n][capacity]
    selected_items = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:
            selected_items.append(i - 1)
            w = int(w - weights[i - 1])

    # Total uang yang digunakan
    total_cost = sum(weights[i] for i in selected_items)
    # Menghitung rata-rata rating produk yang dipilih
    average_rating = total_profit / len(selected_items) if selected_items else 0

    return total_profit, average_rating, total_cost, selected_items


# Load dataset
data = pd.read_csv('/content/cleaned_dataset.csv')

# Menyiapkan data untuk masalah Knapsack
weights = data['Price'].tolist()  # Harga sebagai berat
profits = data['Rating'].tolist()  # Rating sebagai nilai
capacity = 200  # Anggaran maksimal

# Mulai menghitung waktu eksekusi
start_time = time.time()

# Menjalankan fungsi Knapsack
max_value, avg_rating, total_cost, selected_items = knapsackDP(weights, profits, capacity)

# Menghitung waktu eksekusi dan operasi
end_time = time.time()
execution_time = end_time - start_time
operations = len(weights) * capacity

# Menampilkan hasil
print(f"Max Value (Total Rating): {max_value}")
print(f"Average Rating of Selected Products: {avg_rating}")
print(f"Total Cost Used: {total_cost}")
print(f"Execution Time: {execution_time:.4f} seconds")
print(f"Number of Operations: {operations}")

# Kompleksitas Waktu dan Ruang
print("\nComplexity Analysis:")
# Replacing 'items' with 'weights' and 'budget' with 'capacity' as they represent the respective values
print(f"Time Complexity: O(n * W) = O({len(weights)} * {capacity})")
print(f"Space Complexity: O(n * W) = O({len(weights)} * {capacity})")


# Menampilkan item yang dipilih
print("\nSelected Products:")
for i in selected_items:
    print(f"- {data['Title'].iloc[i]} | Price: {data['Price'].iloc[i]} | Rating: {data['Rating'].iloc[i]}")
```

Penyelesaian Permasalahan Integer Knapsack menggunakan Branch and Bound

```python
import pandas as pd
import time

# Node class untuk menyimpan informasi dalam pohon pencarian
class Node:
    def __init__(self, level, profit, weight, bound, include=None):
        self.level = level  # Level dalam pohon (seberapa banyak item yang sudah dipilih)
        self.profit = profit  # Profit (nilai) yang sudah diperoleh
        self.weight = weight  # Berat yang sudah dipilih
        self.bound = bound  # Batas nilai maksimum yang bisa dicapai
        self.include = include if include is not None else []  # Menyimpan status item yang dipilih

# Fungsi untuk menghitung batas atas (bound) untuk cabang tertentu
def bound(node, n, W, weights, profits):
    if node.weight >= W:
        return 0  # Jika berat melebihi kapasitas, tidak ada keuntungan lebih
    profit_bound = node.profit
    j = node.level + 1
    total_weight = node.weight

    # Menambahkan item-item berikutnya dengan pendekatan Greedy (fractional)
    while j < n and total_weight + weights[j] <= W:
        total_weight += weights[j]
        profit_bound += profits[j]
        j += 1

    # Menambahkan sebagian item terakhir jika memungkinkan
    if j < n:
        profit_bound += (W - total_weight) * profits[j] / weights[j]

    return profit_bound

# Fungsi Branch and Bound untuk Knapsack
def knapsackBranchAndBound(weights, profits, capacity):
    n = len(weights)

    # Priority queue untuk menyimpan nodes yang akan diproses
    Q = []

    # Membuat node awal (level = -1, profit = 0, berat = 0)
    root = Node(-1, 0, 0, 0)
    root.bound = bound(root, n, capacity, weights, profits)

    # Menambahkan root ke dalam antrian
    Q.append(root)
    max_profit = 0
    selected_items = []

    while Q:
        # Mengambil node dengan profit tertinggi dari antrian
        node = Q.pop(0)

        # Jika node sudah berada pada level terakhir, lanjutkan
        if node.level == n - 1:
            continue

        # Membuat dua cabang untuk memasukkan atau tidak memasukkan item berikutnya
        level = node.level + 1

        # Node ketika item tidak dimasukkan
        left = Node(level, node.profit, node.weight, 0, node.include.copy())
        left.bound = bound(left, n, capacity, weights, profits)
        if left.bound > max_profit:  # Hanya masukkan jika bound lebih baik
            Q.append(left)

        # Node ketika item dimasukkan
        right = Node(level, node.profit + profits[level], node.weight + weights[level], 0, node.include.copy())
        right.include.append(level)  # Tandai item ini sebagai dipilih
        if right.weight <= capacity:
            right.bound = bound(right, n, capacity, weights, profits)
            if right.bound > max_profit:
                Q.append(right)
                if right.profit > max_profit:
                    max_profit = right.profit
                    selected_items = right.include  # Menyimpan item yang dipilih

    return max_profit, selected_items

# Memuat dataset (contoh)
data = pd.read_csv('/content/cleaned_dataset.csv')

# Menyiapkan data untuk masalah Knapsack
weights = data['Price'].tolist()  # Harga sebagai berat
profits = data['Rating'].tolist()  # Rating sebagai nilai
capacity = 200  # Anggaran maksimal

# Menjalankan Branch and Bound
start_time = time.time()
max_value, selected_items = knapsackBranchAndBound(weights, profits, capacity)
end_time = time.time()

# Menghitung waktu eksekusi dan operasi
execution_time = end_time - start_time
operations = len(weights) * capacity

# Menghitung total biaya dan rata-rata rating
total_cost = sum(weights[i] for i in selected_items)
average_rating = max_value / len(selected_items) if selected_items else 0

# Menampilkan hasil
print(f"Nilai Maksimal (Total Rating): {max_value}")
print(f"Rata-rata Rating Produk yang Dipilih: {average_rating}")
print(f"Total Biaya yang Digunakan: {total_cost}")
print(f"Waktu Eksekusi: {execution_time:.4f} detik")
print(f"Jumlah Operasi: {operations}")

# Kompleksitas Waktu dan Ruang
print("\nAnalisis Kompleksitas:")
print(f"Kompleksitas Waktu: O(n * W) = O({len(weights)} * {capacity})")
print(f"Kompleksitas Ruang: O(n * W) = O({len(weights)} * {capacity})")

# Menampilkan item yang dipilih
print("\nProduk yang Dipilih:")
for i in selected_items:
    print(f"- {data['Title'].iloc[i]} | Harga: {data['Price'].iloc[i]} | Rating: {data['Rating'].iloc[i]}")
```
